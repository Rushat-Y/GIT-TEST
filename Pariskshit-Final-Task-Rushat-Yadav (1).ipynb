```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
```

```python
image_path = 'Copy of LC08_L1TP_011019_20240102_20240102_02_RT_TIR.jpg'
im1 = cv2.imread(image_path)
im1_rgb = cv2.cvtColor(im1, cv2.COLOR_BGR2RGB)
plt.imshow(im1_rgb)
plt.axis('off')
plt.title("Image Display")
plt.show()
```

```python
(h, w) = im1.shape[:2]
center = (w // 2, h // 2)
angle = 17.5
M = cv2.getRotationMatrix2D(center, angle, 1.0)
rotated = cv2.warpAffine(im1, M, (w, h))
plt.imshow(rotated)
plt.axis('off') 
plt.title("Image Display")
plt.show()
```

```python
cv2.imshow('image window',rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotated_r = cv2.resize(rotated, (356,360))
```

```python
cv2.imshow('image window',rotated_r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_img=cv2.cvtColor(rotated_r, cv2.COLOR_BGR2GRAY)
histeq = cv2.equalizeHist(gray_img)
res = np.hstack((gray_img, histeq))
plt.imshow(res)
plt.axis('off') 
plt.title("Image Display")
plt.show()
```

```python
cv2.imshow('image', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(res,cmap='gray')
plt.axis('off') 
plt.title("Image Display")
plt.show()
```

```python
all_img = np.hstack((rotated_r, rotated_r2,rotated_r3,rotated_r4,rotated_r5))
plt.imshow(all_img)
plt.axis('off') 
plt.title("All Images")
plt.show()
```

```python
cv2.imshow('image window',all_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_all_img=cv2.cvtColor(all_img, cv2.COLOR_BGR2GRAY)
histeq_all = cv2.equalizeHist(gray_all_img)
plt.imshow(histeq_all,cmap='gray')
plt.axis('off') 
plt.title("All Images HISTEQ")
plt.show()
```

```python
cv2.imshow('image window',histeq_all)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img_all = clahe.apply(gray_all_img)
plt.imshow(clahe_img_all,cmap='gray')
plt.axis('off') 
plt.title("All Images CLAHE")
plt.show()
plt.imshow(histeq_all,cmap='gray')
plt.axis('off') 
plt.title("All Images HISTEQ")
plt.show()
```

```python
cv2.imshow('image window',clahe_img_all)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
smoothed_img = cv2.GaussianBlur(clahe_img_all, (5, 5), 0)
plt.imshow(smoothed_img,cmap='gray')
plt.axis('off') 
plt.show()
cv2.imshow('image window',smoothed_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
